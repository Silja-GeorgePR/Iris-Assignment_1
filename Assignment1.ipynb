# 1. Read the dataset to python environment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_set= pd.read_excel("iris.xls")
data_set
```

# 2. Display the columns in the dataset.

```python
data_set.head()
```

# 3. Calculate the mean of each column of the dataset.

```python

numerical_column=data_set.select_dtypes(include="number")

mean_of_columns = numerical_column.mean()
print("Mean of SL,SW,PL,PW is: ")
mean_of_columns
```

# 4. Check for the null values present in the dataset.

```python

is_null=data_set.isnull().values.any()
print(is_null)
count_null=data_set.isnull().values.any().sum()
print(count_null)
```

# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.


```python
for classes in data_set['Classification'].unique():
    diff_classification = data_set[data_set["Classification"] == classes]
   
    columns= diff_classification.columns[0:4]
    for i in columns:
        sns.histplot(diff_classification[i],color="green")
        plt.title(classes)
       
        plt.show()
```

```python
sns.lineplot(x=data_set['SL'],y=data_set['PL'])
plt.title("Correlation between Petal and Sepal length")
plt.show()
correlation= data_set['SL'].corr(data_set['PL'])
correlation
```

```python

irisVerColor= data_set.loc[(data_set['Classification']== "Iris-versicolor")]
irisSet=data_set.loc[(data_set['Classification']== "Iris-setosa")]
irisVir=data_set.loc[(data_set['Classification']== "Iris-virginica")]
fig, axes = plt.subplots(1, 3, figsize=(10, 7), sharey=True)
fig.suptitle('Distribution of each attribute values')

irisVerColor.boxplot(ax=axes[0])
axes[0].set_title("Iris-Versicolor")
irisSet.boxplot(ax=axes[1])
axes[1].set_title(" Iris-Setosa")
irisVir.boxplot(ax=axes[2])
axes[2].set_title(" Iris-Virginica")


```
